```python
from halotools.empirical_models import PrebuiltHodModelFactory
zehavi11_model = PrebuiltHodModelFactory('zheng07', modulate_with_cenocc=True)
```

```python
zehavi11_model.param_dict['sigma_logM'] = 0.25
zehavi11_model.param_dict['logM0'] = 12.35
zehavi11_model.param_dict['logMmin'] = 11.83
zehavi11_model.param_dict['alpha'] = 1.00
zehavi11_model.param_dict['logM1'] = 13.08
```

```python
from astropy.table import Table
ab_model_fname='best_fit_hod_zentner_model.hdf5'
std_model_fname='best_fit_hod_zheng_model.hdf5'
ab_model = Table.read(ab_model_fname, path='data')
std_model = Table.read(std_model_fname, path='data')
```

```python
plt.plot(ab_model['mass_bin_midpoints'], ab_model['mean_ncen'], 
         color='red', label='assembly-biased model')
plt.plot(std_model['mass_bin_midpoints'], std_model['mean_ncen'], 
         color='blue', label='standard model')

mass_array = np.logspace(10, 15, 100)
plt.plot(mass_array, zehavi11_model.mean_occupation_centrals(prim_haloprop=mass_array), 
         '--', color='k', label='Zehavi+11')

plt.loglog()
plt.xticks(size=15); plt.yticks(size=15)
plt.xlim(xmin = 1e11, xmax=1e15)
plt.ylim(ymin = 0.05, ymax = 1.1)
plt.xlabel(r'$M_{\rm vir}$  $[M_{\odot}/h]$', size=20)
plt.ylabel(r'$\langle N_{\rm cen} \rangle$', size=20)

plt.legend(fontsize=15, frameon=False, loc='best')
plt.savefig('central_hod_comparison.pdf')
```

```python
plt.plot(ab_model['mass_bin_midpoints'], ab_model['mean_nsat'], 
         color='red', label='assembly-biased model')
plt.plot(std_model['mass_bin_midpoints'], std_model['mean_nsat'], 
         color='blue', label='standard model')

mass_array = np.logspace(10, 15, 100)
plt.plot(mass_array, zehavi11_model.mean_occupation_satellites(prim_haloprop=mass_array), 
         '--', color='k', label='Zehavi+11')

plt.loglog()
plt.xticks(size=15); plt.yticks(size=15)
plt.xlim(xmin = 1e12, xmax=1e15)
plt.ylim(ymin = 0.01, ymax = 50)
plt.xlabel(r'$M_{\rm vir}$  $[M_{\odot}/h]$', size=20)
plt.ylabel(r'$\langle N_{\rm sat} \rangle$', size=20)

plt.legend(fontsize=15, frameon=False, loc='best')
plt.savefig('satellite_hod_comparison.pdf')
```
