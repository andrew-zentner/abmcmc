```python
%matplotlib inline
from matplotlib import ticker
```

```python
from halotools.empirical_models import PrebuiltHodModelFactory
zehavi11_model = PrebuiltHodModelFactory('zheng07', modulate_with_cenocc=True)
```

```python
zehavi11_model.param_dict['sigma_logM'] = 0.25
zehavi11_model.param_dict['logM0'] = 12.35
zehavi11_model.param_dict['logMmin'] = 11.83
zehavi11_model.param_dict['alpha'] = 1.00
zehavi11_model.param_dict['logM1'] = 13.08
```

```python
from astropy.table import Table
ab_model_fname='best_fit_hod_zentner_model.hdf5'
std_model_fname='best_fit_hod_zheng_model.hdf5'
ab_model = Table.read(ab_model_fname, path='data')
std_model = Table.read(std_model_fname, path='data')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(ab_model['mass_bin_midpoints'], ab_model['mean_ncen'], 
         color='forestgreen', label='assembly-biased model', linewidth=2)
ax.plot(std_model['mass_bin_midpoints'], std_model['mean_ncen'], 
         color='crimson', label='standard model', linewidth=2)

# Plot errors
y1 = ab_model['mean_ncen'] - ab_model['stddev_ncen']
y2 = ab_model['mean_ncen'] + ab_model['stddev_ncen']
y_low = np.where(y2 < y1, y2, y1)
y_high = np.where(y2 > y1, y2, y1)

ax.fill_between(ab_model['mass_bin_midpoints'], y_low,  y_high, 
                where = y2 >= y1, facecolor='lightgreen', color="none", interpolate=True)

mass_array = np.logspace(10, 15, 100)
ax.plot(mass_array, zehavi11_model.mean_occupation_centrals(prim_haloprop=mass_array), 
         '--', color='k', label='Zehavi+11', linewidth=2)

plt.loglog()
plt.xticks(size=15); plt.yticks(size=15)
plt.xlim(xmin = 1e11, xmax=5e14)
plt.ylim(ymin = 0.05, ymax = 1.1)
xlabel = ax.set_xlabel(r'$M_{\rm vir}$  $[M_{\odot}/h]$', size=23)
ylabel = ax.set_ylabel(r'$\langle N_{\rm cen} \rangle$', size=23)

ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.1f'))


plt.legend(fontsize=15, frameon=False, loc='best')
fig.savefig('central_hod_comparison.pdf', 
                  bbox_extra_artists=[xlabel, ylabel], bbox_inches='tight')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(ab_model['mass_bin_midpoints'], ab_model['mean_nsat'], 
         color='forestgreen', label='assembly-biased model', linewidth=2)
ax.plot(std_model['mass_bin_midpoints'], std_model['mean_nsat'], 
         color='crimson', label='standard model', linewidth=2)

mass_array = np.logspace(10, 15, 100)
ax.plot(mass_array, zehavi11_model.mean_occupation_satellites(prim_haloprop=mass_array), 
         '--', color='k', label='Zehavi+11', linewidth=2)

# Plot errors
y1 = ab_model['mean_nsat'] - ab_model['stddev_nsat']
y2 = ab_model['mean_nsat'] + ab_model['stddev_nsat']
y_low = np.where(y2 < y1, y2, y1)
y_high = np.where(y2 > y1, y2, y1)
ax.fill_between(ab_model['mass_bin_midpoints'], y_low, y_high,
                facecolor='lightgreen', color="none", interpolate=True)

plt.loglog()
plt.xticks(size=15); plt.yticks(size=15)
plt.xlim(xmin = 1e12, xmax=1e15)
plt.ylim(ymin = 0.01, ymax = 50)
xlabel = ax.set_xlabel(r'$M_{\rm vir}$  $[M_{\odot}/h]$', size=23)
ylabel = ax.set_ylabel(r'$\langle N_{\rm sat} \rangle$', size=23)

ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.1f'))

plt.legend(fontsize=15, frameon=False, loc='best')
fig.savefig('satellite_hod_comparison.pdf', 
                  bbox_extra_artists=[xlabel, ylabel], bbox_inches='tight')
```
